```python
# Dependencies
import pandas as pd

```

```python
# List of file names to be combined

d1 = pd.read_csv("Resources/201307-citibike-tripdata.csv")
d2 = pd.read_csv("Resources/201308-citibike-tripdata.csv")
d3 = pd.read_csv("Resources/201309-citibike-tripdata.csv")
d4 = pd.read_csv("Resources/201310-citibike-tripdata.csv")
d5 = pd.read_csv("Resources/201311-citibike-tripdata.csv")
d6 = pd.read_csv("Resources/201312-citibike-tripdata.csv")
d7 = pd.read_csv("Resources/201401-citibike-tripdata.csv")
d8 = pd.read_csv("Resources/201402-citibike-tripdata.csv")
d9 = pd.read_csv("Resources/201306-citibike-tripdata.csv")
```

```python
# Concatenate all data to one dataframe

combined_df = pd.concat([d1,d2,d3,d4,d5,d6,d7,d8,d9])
combined_df.count()
```

```python
# Dictionary mapping values in the 'gender' column to new values
mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}

# Use the 'map' method to change the values in the 'gender' column
combined_df['gender'] = combined_df['gender'].map(mapping)

# Show DataFrame
combined_df
```

```python
# Convert the 'date' column to a datetime object, handling errors
combined_df['starttime'] = pd.to_datetime(combined_df['starttime'], errors='coerce')

# Extract the year and create a new column
combined_df['year'] = combined_df['starttime'].dt.year
```

```python
# Extract the month and create a new column
combined_df['month'] = combined_df['starttime'].dt.strftime('%B')
```

```python
# Show dataframe
combined_df
```

```python
# Save the dataframe to a CSV file

combined_df.to_csv('Resources/citibike_tripdata.csv', index=False)
```
